```python
import json

def out(file):
    output = {}
    text = open(file).read()
    output['title'] = file
    output['text'] = text
    open("%s.json"%file,'w').write(json.dumps(output))

out('d1.txt')
out('d2.txt')
out('d3.txt')
out('d4.txt')
out('d5.txt')
out('d6.txt')
```

```python
import pandas as pd
domains_embeddings = pd.read_csv('domains_embeddings.csv',header=None,skiprows=1).set_index(0)
results_embeddings = pd.read_csv('results_embeddings.csv',header=None,skiprows=1).set_index(0)
domains_embeddings
```

```python
results_embeddings
```

```python
from sklearn.metrics.pairwise import cosine_similarity

sims = pd.DataFrame(cosine_similarity(results_embeddings,domains_embeddings),index=results_embeddings.index,columns=domains_embeddings.index)
sims.iloc[0].sort_values(ascending=False).index[:3]
```
