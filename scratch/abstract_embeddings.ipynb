```python
import pandas as pd
```

```python
df = pd.read_csv('embeddings.csv',skiprows=1,header=None)
df.columns = ["uri"]+[f"x{i}" for i in range(df.shape[1]-1)]
df = df.set_index('uri')
df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
sim_df = pd.DataFrame(cosine_similarity(df),columns=df.index,index=df.index)
sim_df.values[np.tril_indices(sim_df.shape[0], 0)] = np.nan
stacked_sim_df = sim_df.stack().sort_values().to_frame()
stacked_sim_df.columns = ['Cosine Similarity']
stacked_sim_df
```

```python
stacked_sim_df.to_csv('abstracts_stacked_sim_df.csv')
```
