```python
import pandas as pd
results = pd.read_csv('tmp/1.38037-PB1-9531-R2.pdf.tika.cleaned.csv',index_col=0)
results
```

```python
good_results = results[results['Score']<=300]
good_results
```

```python
good_paragraphs = good_results.groupby('Paragraph ID').apply(lambda x: ".".join(x['Sentence']))
good_paragraphs = good_paragraphs[good_paragraphs.str.len() > 50] # min num characters
good_paragraphs
```

```python
i = 0
for p in good_paragraphs:
    print(i,len(p))
    open(f"tmp/content_{i}.txt","w").write(str(p))
    i+=1
```
